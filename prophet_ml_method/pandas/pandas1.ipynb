#### first we need x axis and y axis to visualize the data and date must be the the x axis and the data will be the y axis

```python
start=datetime.datetime(2010,1,1) #starting date from where we start
end=datetime.datetime.now()
#print start
#print end
```

```python
#create a data frame thats what pandas do
df = web.DataReader('XOM',"morningstar",start,end)
```

#### we pull the data from the morningstar stock price and create a data frame for further calculation

```python
#now print some of the initial element we got in the dataframe so see whats we are up to
df.head()
```

#### just like previously said we gonna take the date as a X so we have to set it as a index and remove the XOM thing it is not necessary

```python
df.reset_index(inplace=True)
df.head()
```

#### as we can see that the index has been reset everything become a label with feature so now we have to delete the label with the XOM and set the the date as an index

```python
df = df.drop('Symbol',axis=1)
df.head()
```
