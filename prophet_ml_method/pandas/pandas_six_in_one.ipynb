```python
## so togather we will write a function
## in oop format
import datetime
import pandas
import pandas_datareader.data as web
import matplotlib.pyplot as plt
from matplotlib import style
class data:
    ##Idefining the constructor

    def get_the_data(self):
        self.start=datetime.datetime(2010,1,1)
        self.end=datetime.datetime.now()
        self.df=web.DataReader("XOM","morningstar",self.start,self.end)
        return self.df
    def data_manu(self):
        ## reset thge index
        self.df.reset_index(inplace=True)
        ##remove the first label
        self.df=self.df.drop('Symbol',axis=1)
        self.df.set_index("Date",inplace=True)
    def print_data(self):
        return self.df
    def plotting_the_data(self):
        return self.df['High'].plot()
        
```
