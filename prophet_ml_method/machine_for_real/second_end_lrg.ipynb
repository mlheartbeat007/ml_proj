```python
def decision(accuracy1,accuracy2,accracy3):
    data=[accuracy1,accuracy2,accuracy3]
    result=max(data)
    if result == accuracy1:
        method = "KNN"
        
    elif result==accuracy2:
        method='LR'
    else:
        method='SVR'
    return method
```

```python
best_method=decision(accuracy1,accuracy2,accuracy3)
```
