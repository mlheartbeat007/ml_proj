# multibar graph
#### in the multi bar graph we have to do something  more first as the Y axis we have to assaign list insted of integer for every component of the x axis so Y axis will be a multi-dimensional list and we have to do one more thing we have to shift the every list plot more than their width

```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint as rint # normal aliasing 
#this random module will actually genarate random integer
# so everytime we run the program it will genarate new value
```

```python
#lets do some loop to assign the 4 list 
# and then create a multidimensonal loop
# there will be 4 element of the each loop
a=[]
b=[]
c=[]
d=[]
for item in range(4):
    a.append(rint(20,50)) #$range of the random number
    b.append(rint(20,50))
    c.append(rint(20,50))
    d.append(rint(20,50))

y=[a,b,c,d]
#now we create a x range.we can do with range() function
# we will use the numpy module
x=np.arange(4) # the number will be the element of the list
#inside the list
#it creates an array.np allow us to use array
```

#### plotting command starts

```python
plt.bar(x+0.00,y[0],width=0.20)
plt.bar(x+0.25,y[1],width=0.20)
plt.bar(x+0.50,y[2],width=0.20)
```

```python
plt.show() # it is necessary if you are not using the note book
# and you have to also change the color manually
```

# back to back graph

```python
# now we need numpy for sure 
# to create a array because python does not support array
# numpy will allow us to use an array
#first import everything
import numpy as np     # aliasing
import matplotlib.pyplot as plt 
```

# what is a back to back graph?
#### in back to back graph.the graph starts form the middle and we need two dets of value. whic is the Y axis.for make the back to back graph work we need to take the negatice value of one of the list.and array array length must have to be same and in back to back graph x is our traditional y axis vice versa .and we compare each of the value of those list

```python
#create two sets of array
y1=np.array([10,25,30,35]) #first array 
y2=np.array([15,35,45,20]) #second array
```

```python
# create the X axis 
# X axis will deter mine how many bar will be genarated
x=np.arange(4)   #works like the range function
```

# barh() --> back to bach graph command
## barh ->bar horijontal

```python
plt.barh(x,y1)
plt.barh(x,-y2)
```

```python
plt.show()
```
